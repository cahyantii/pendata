# **FUZZY C-MEANS CLUSTERING**

Fuzzy C-Means adalah cara mengelompokkan data yang mirip dengan K-Means, tetapi lebih fleksibel. Jika K-Means hanya membolehkan satu data masuk ke satu kelompok, Fuzzy C-Means memperbolehkan satu data berada di beberapa kelompok sekaligus. Tiap data punya nilai keanggotaan untuk tiap kelompok, tergantung seberapa dekat data itu dengan pusat kelompoknya

Berbeda dengan K-Means yang menggunakan pendekatan hard clustering (pengelompokan tegas), Fuzzy C-Means (FCM) menerapkan konsep soft clustering, di mana setiap data dapat memiliki keanggotaan pada lebih dari satu cluster secara bersamaan. Pendekatan ini lebih sesuai untuk data yang memiliki karakteristik samar atau batas antar kelompok yang tidak jelas, karena memberikan fleksibilitas dalam menentukan tingkat kedekatan suatu data terhadap setiap pusat cluster.

## FUZZY C-MEANS CLUSTERING

Fuzzy C-Means Clustering merupakan salah satu metode pengelompokan data yang memiliki kemiripan dengan K-Means, namun menawarkan pendekatan yang lebih fleksibel dalam menentukan keanggotaan data terhadap cluster. Berbeda dengan K-Means yang menetapkan setiap data hanya masuk ke satu cluster secara pasti, Fuzzy C-Means memungkinkan setiap data memiliki tingkat keanggotaan (membership degree) pada seluruh cluster yang ada. Dengan kata lain, satu data dapat tergolong ke beberapa cluster sekaligus, dengan nilai keanggotaan tertentu yang mencerminkan kedekatannya terhadap masing-masing pusat cluster (centroid)

Tidak seperti K-Means yang menerapkan pendekatan hard clustering di mana setiap data hanya dimasukkan ke satu cluster secara mutlak, Fuzzy C-Means (FCM) menggunakan pendekatan soft clustering, yang memungkinkan data memiliki keanggotaan di beberapa cluster sekaligus. Oleh karena itu, metode ini lebih sesuai diterapkan pada data yang memiliki batas antar kelompok yang kabur atau tidak terdefinisi secara jelas.

### KONSEP DASAR FUZZY C-MEANS CLUSTERING

Konsep dasar dari Fuzzy C-Means Clustering terletak pada penggunaan derajat keanggotaan (dikenal sebagai membership values) yang menunjukkan sejauh mana suatu data menjadi bagian dari masing-masing cluster. Nilai keanggotaan ini berada dalam kisaran 0 sampai 1, dan untuk setiap data, total jumlah keanggotaannya terhadap seluruh cluster harus sama dengan 1.

Dalam FCM, terdapat beberapa komponen penting:

* Data Point: Objek-objek yang akan diklasterkan.
* Centroid: Titik pusat dari setiap cluster.
* Fuzzy Membership Matrix (U): Matriks yang berisi nilai keanggotaan dari setiap data terhadap setiap cluster.
* Parameter Fuzziness (m): Nilai konstanta yang mengatur tingkat keburaman (biasanya m = 2). Semakin besar nilai m, semakin kabur keanggotaan suatu data pada cluster.


### ALGORITMA FUZZY C-MEANS CLUSTERING

Proses kerja algoritma Fuzzy C-Means adalah sebagai berikut:

1. Inisialisasi:

   * Tentukan jumlah cluster (*c*), parameter fuzziness (*m*), batas toleransi error (*Îµ*), dan nilai maksimum iterasi.
   * Inisialisasi *membership matrix* $U_{ij}$ secara acak, dengan syarat:

     $$
     \sum_{j=1}^c U_{ij} = 1, \quad \forall i
     $$

2. Hitung Centroid Tiap Cluster:

   $$
   v_j = \frac{\sum_{i=1}^n (U_{ij})^m \cdot x_i}{\sum_{i=1}^n (U_{ij})^m}
   $$

   Di mana:

   * $v_j$ adalah pusat cluster ke-j
   * $x_i$ adalah data ke-i
   * $U_{ij}$ adalah nilai keanggotaan data ke-i terhadap cluster ke-j

3. Update Membership Matrix:

   $$
   U_{ij} = \frac{1}{\sum_{k=1}^c \left( \frac{\|x_i - v_j\|}{\|x_i - v_k\|} \right)^{\frac{2}{m-1}}}
   $$

   Di mana:

   * $\|x_i - v_j\|$ adalah jarak antara data ke-i dengan pusat cluster ke-j

4. Pengecekan Konvergensi:

   * Jika perubahan nilai *membership matrix* kurang dari Îµ atau sudah mencapai iterasi maksimum, maka proses dihentikan.
   * Jika belum, kembali ke langkah 2.

## CONTOH MANUAL PERHITUNGAN FUZZY C-MEANS CLUSTERING
Berikut adalah versi **contoh manual perhitungan Fuzzy C-Means Clustering** yang sama persis strukturnya, namun **angka-angkanya telah diubah** secara konsisten. Misalnya, kita ubah dataset awal menjadi:

**Data:**

Misalnya kita memiliki 4 data satu dimensi:

$$
X = \{3, 4, 8, 9\}
$$

**Jumlah Cluster (c):** 2
**Tingkat Fuzziness (m):** 2
**Kriteria Konvergensi:** perubahan pusat cluster < 0.01

---

### **Langkah 1: Inisialisasi Matriks Keanggotaan \$U\$**

Ukuran matriks: \$2 \times 4\$

Misalkan kita mulai dari:

$$
U^{(0)} =
\begin{bmatrix}
0.7 & 0.6 & 0.3 & 0.2 \\
0.3 & 0.4 & 0.7 & 0.8 \\
\end{bmatrix}
$$

Sudah valid karena jumlah tiap kolom = 1.

---

### **Langkah 2: Hitung Pusat Cluster (Centroid)**

Rumus:

$$
v_j = \frac{\sum u_{ji}^m \cdot x_i}{\sum u_{ji}^m}
$$

#### Cluster 1:

$$
v_1 = \frac{(0.7^2 \cdot 3) + (0.6^2 \cdot 4) + (0.3^2 \cdot 8) + (0.2^2 \cdot 9)}{0.7^2 + 0.6^2 + 0.3^2 + 0.2^2}
= \frac{1.47 + 1.44 + 0.72 + 0.36}{0.49 + 0.36 + 0.09 + 0.04}
= \frac{4.0}{0.98}
\approx 4.082
$$

#### Cluster 2:

$$
v_2 = \frac{(0.3^2 \cdot 3) + (0.4^2 \cdot 4) + (0.7^2 \cdot 8) + (0.8^2 \cdot 9)}{0.3^2 + 0.4^2 + 0.7^2 + 0.8^2}
= \frac{0.27 + 0.64 + 3.92 + 5.76}{0.09 + 0.16 + 0.49 + 0.64}
= \frac{10.59}{1.38}
\approx 7.672
$$

---

### **Langkah 3: Update Matriks Keanggotaan \$U\$**

Contoh: untuk data \$x\_1 = 3\$

* \$d(x\_1, v\_1) = |3 - 4.082| = 1.082\$
* \$d(x\_1, v\_2) = |3 - 7.672| = 4.672\$

$$
u_{1,1} = \frac{1}{1 + \left(\frac{1.082}{4.672}\right)^2}
= \frac{1}{1 + 0.0536}
= \frac{1}{1.0536}
\approx 0.949
$$

$$
u_{2,1} = 1 - u_{1,1} \approx 0.051
$$

Ulangi untuk semua data dan semua cluster.

---

### **Langkah 4: Cek Konvergensi**

Cek perubahan nilai centroid. Jika semua \$\Delta v\_j < 0.01\$, maka **konvergen**.

---

### **Fungsi Objektif \$J\_m\$**

$$
J_m = \sum_{j=1}^{c} \sum_{i=1}^{n} u_{ji}^m \cdot (x_i - v_j)^2
$$

Lanjutkan dengan nilai \$u\_{ji}\$ dan \$v\_j\$ dari hasil sebelumnya.

## IMPLEMENTASI PYTHON

Berikut juga saya lakukan visualisasi 3 Dimensi pada data Iris setelah clustering

```python
import numpy as np
import pandas as pd

# Data dan parameter
X = np.array([3, 4, 8, 9])  # Data satu dimensi
c = 2                      # Jumlah cluster
m = 2.0                    # Tingkat fuzziness
epsilon = 0.01             # Ambang batas konvergensi
max_iter = 100             # Maksimum iterasi

# Inisialisasi keanggotaan U secara manual (2 x 4)
U = np.array([
    [0.7, 0.6, 0.3, 0.2],
    [0.3, 0.4, 0.7, 0.8]
])

# Fungsi untuk menghitung pusat cluster
def calculate_centroids(U, X, m):
    um = U ** m
    return (um @ X) / um.sum(axis=1)

# Fungsi untuk update U
def update_membership(X, V, m):
    n = len(X)
    c = len(V)
    U_new = np.zeros((c, n))
    for i in range(n):
        for j in range(c):
            denom_sum = 0.0
            for k in range(c):
                dist_ratio = abs(X[i] - V[j]) / abs(X[i] - V[k]) if abs(X[i] - V[k]) != 0 else 1e-10
                denom_sum += dist_ratio ** (2 / (m - 1))
            U_new[j, i] = 1.0 / denom_sum
    return U_new

# Fungsi objektif
def objective_function(U, V, X, m):
    obj = 0.0
    for j in range(c):
        for i in range(len(X)):
            obj += (U[j, i] ** m) * (abs(X[i] - V[j]) ** 2)
    return obj

# Iterasi sampai konvergen
history = []
for iteration in range(max_iter):
    V = calculate_centroids(U, X, m)
    obj = objective_function(U, V, X, m)
    history.append((iteration+1, V.copy(), U.copy(), obj))

    U_new = update_membership(X, V, m)

    if np.max(np.abs(V - calculate_centroids(U_new, X, m))) < epsilon:
        break
    U = U_new

# Konversi hasil ke DataFrame agar mudah dibaca
results = []
for iter_num, centroids, membership, obj_val in history:
    result = {
        "Iterasi": iter_num,
        "Centroid 1": centroids[0],
        "Centroid 2": centroids[1],
        "Objektif": obj_val
    }
    for j in range(c):
        for i in range(len(X)):
            result[f"U[{j+1},{i+1}]"] = membership[j, i]
    results.append(result)

df_results = pd.DataFrame(results)
df_results.round(4)  # Pembulatan 4 desimal untuk keterbacaan

```

## Clustering menggunakan Fuzzy C-Means

### Clustering dengan 2 cluster dan Visualisasi data

Disini saya melakukan clustering menggunakan Fuzzy C-Means dengan 2 cluster sekaligus menghitung Pusat Cluster dan Fungsi Objektif dari semua iterasi, lalu melakukan visualisasi 2 Dimensi pada data Iris setelah clustering

```python

# Transpose fitur agar sesuai dengan input skfuzzy (fitur x sample)
data = scaled_features.T

# Fuzzy C-Means clustering
n_clusters = 2
cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
    data, c=n_clusters, m=2.0, error=0.005, maxiter=1000, init=None, seed=42
)

# Ambil label keras dari keanggotaan tertinggi
labels_2 = np.argmax(u, axis=0)

# Evaluasi clustering
silhouette_fuzzy = silhouette_score(scaled_features, labels_2)

# Cetak hasil evaluasi
print(f"[Fuzzy C-Means - 2 Clusters]")
print(f"Silhouette Score: {silhouette_fuzzy:.4f}")
print(f"Fuzzy Partition Coefficient (FPC): {fpc:.4f}")
print(f"Objective Function Value (Jm): {jm[-1]:.4f}")  # ambil nilai terakhir iterasi
print(f"Cluster Centers (in scaled feature space):\n{cntr}")

# Menampilkan nilai fungsi objektif di setiap iterasi
print("Objective Function Values per Iteration:")
for i, val in enumerate(jm):
    print(f"Iterasi {i+1:02d}: {val:.4f}")


# Visualisasi hasil clustering (dalam ruang PCA)
plt.figure(figsize=(6, 4))
plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=labels_2, cmap='viridis', s=50)
plt.title("Fuzzy C-Means Clustering (2 Clusters)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.show()
```
